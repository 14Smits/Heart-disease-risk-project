```python
import pandas as pd
```

```python
df = pd.read_csv("heart-categories.csv")
```

```python
df.head()
```

```python
columns_to_keep=["State","Sex","HadHeartAttack"]
```

```python
columns_to_drop=[col for col in df.columns if col not in columns_to_keep]
df.drop(columns=columns_to_drop,inplace=True)
df['HadHeartAttack'] = df['HadHeartAttack'].map({'Yes': 1, 'No': 0})
df.head()
```

```python
# Filter the DataFrame for positive cases
positive_cases = df[df['HadHeartAttack'] == 1]

# Group by State and Sex and count the occurrences
result = positive_cases.groupby(['State', 'Sex']).size().unstack(fill_value=0).reset_index()

# Rename columns for clarity
result.columns = ['State', 'Female', 'Male']

# Add a 'Total' column
result['Total'] = result['Female'] + result['Male']
#Remove territories and districts to focus only on the US States to simplify our dataset. 
df.drop(df[df['State'].isin(['Guam', 'Virgin Islands',"District of Columbia","Puerto Rico"])].index, inplace=True)

result.to_csv('Tableau_data.csv', index=False)
```
